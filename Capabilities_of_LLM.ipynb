```python
from transformers import pipeline
```

```python
# Sentiment Analysis

classifier = pipeline("sentiment-analysis") # give here the capability of LLM.
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
# Text-Generation
generator = pipeline("text-generation", model="gpt2")

# Generate text
output = generator(
    "In this course, we will teach you how to",
    max_length=50,
    num_return_sequences=1,
    do_sample=True,
    temperature=0.7
)

print(output[0]["generated_text"])
```

```python
# Zero-shot Classification

classifier = pipeline("zero-shot-classification")
classifier("I am working on new start up",
           candidate_labels = ["education", "politics", "business"])
```

```python
!pip install sacremoses # Install recommended dependency

# Language Translation
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "Helsinki-NLP/opus-mt-fr-en"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = "Bonjour tout le monde"

# Encode the input text
input_ids = tokenizer(text, return_tensors="pt").input_ids

# Generate the translation
outputs = model.generate(input_ids)

# Decode the output tokens to get the translated text
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(translated_text)
```

```python
from transformers import pipeline

# Question Answering

qa_pipeline = pipeline("question-answering")

context = """
Hugging Face is a company based in New York City.
It is known for developing the Transformers library.
"""

question = "Where is Hugging Face based?"

result = qa_pipeline(question=question, context=context)

print("Answer:", result["answer"])
print("Score:", result["score"])
```

```python
from transformers import pipeline
# fill the sentence

fill_mask = pipeline("fill-mask", model="bert-base-uncased")

mask_token = fill_mask.tokenizer.mask_token

sentence = f"Machine learning is {mask_token}."

result = fill_mask(sentence)

for r in result:
    print(r["sequence"])
```

```python
from transformers import pipeline
# Name Entity Relationship
ner = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)

text = "Elon Musk founded SpaceX in 2002 in the United States."

result = ner(text)

for entity in result:
    print(f"{entity['word']} --> {entity['entity_group']}")
```
